```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Creating the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# Creating a dictionary with the two lists
movie_dict = {"years":years, "durations":durations}

# Printing the dictionary
print(movie_dict)
```

```python
durations_df = pd.DataFrame(movie_dict)

# Printing the DataFrame
print(durations_df)
```

```python
fig = plt.figure()

# Drawing a line plot of release_years and durations
plt.plot(years, durations)

# Creating a title
plt.title("Netflix Movie Durations 2011-2020")

# Showing the plot
plt.show()
```

```python
# Reading in the CSV as a DataFrame
netflix_df = pd.read_csv("netflix data/netflix_data.csv")

# Printing the first five rows of the DataFrame
print(netflix_df.iloc[0:5])
```

```python
# Subsetting the DataFrame for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df["type"]=='Movie']
print(netflix_df_movies_only)

# Selecting only the columns of interest
netflix_movies_col_subset = netflix_df[['title', 'country', 'genre', 'release_year', 'duration']]


# Printing the first five rows of the new DataFrame
print(netflix_movies_col_subset.iloc[0:5])
```

```python
# Creating a figure and increasing the figure size
fig = plt.figure(figsize=(12,8))

# Creating a scatter plot of duration versus year
list1 = []
list2 = []
for i, j in netflix_movies_col_subset.iterrows():
     list1.append(j['release_year'])
     list2.append(j['duration'])
plt.scatter(list1, list2)
# Creating a title
plt.title("Movie Duration by Year of Release")

# Showing the plot
plt.show()
```

```python
# Filtering for durations shorter than 60 minutes
short_movies = pd.DataFrame(netflix_movies_col_subset[netflix_movies_col_subset['duration']<60])

# Printing the first 20 rows of short_movies
print(short_movies[0:20])
```

```python
# Defining an empty list
colors = []

# Iterating over rows of netflix_movies_col_subset
for lab, row in netflix_movies_col_subset.iterrows():
    if row['genre']=='Children':
        colors.append("red")
    elif row['genre']=='Documentaries':
        colors.append("blue")
    elif row['genre']=='Stand-Up':
        colors.append("green")
    else:
        colors.append("black")

# Inspecting the first 10 values in the list        
print(colors[0:10])
```
